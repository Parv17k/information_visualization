```python
## Line Chart
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
score=np.random.randint(100, size=99)
roll = range(1,100)
plt.title("Students Number")
plt.xlabel('score')
plt.ylabel('roll #')
plt.plot(roll,score)
```

```python
## Bar Chart
import matplotlib.pyplot as plt
import numpy as np
score=np.random.randint(100, size=99)
roll = range(1,100)
plt.title("Students Number")
plt.xlabel('score')
plt.ylabel('roll #')
plt.bar(roll,score)
```
